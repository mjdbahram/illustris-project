```python
import numpy as np
import matplotlib.pyplot as plt
import h5py  
```

```python
x_p = []
y_p = []
z_p = []
file_name = []
for i in range(32):
    file_name.append("./simulation_num_135/snap_135."+str(i)+".hdf5")
    print(i, file_name[-1])
    d1 = h5py.File(file_name[i],"r")
    header_PT0 = d1.get('PartType0')
    x = np.array(header_PT0['Coordinates']).transpose()[0]
    y = np.array(header_PT0['Coordinates']).transpose()[1]
    z = np.array(header_PT0['Coordinates']).transpose()[2]

    name_string = "Chunck_num: "+str(i)
    fig, ax = plt.subplots(1,3,figsize = (12,4));
    fig.suptitle(name_string, y=1.08)

    ax[0].set_xticks(np.arange(0,8e4,1.5e4))
    ax[0].hist(x, bins=20)
    ax[0].set_xlabel("x coordinate in ckpc ")
    ax[0].set_ylabel("num")

    ax[1].set_xticks(np.arange(0,8e4,1.5e4))
    ax[1].hist(y, bins=20)
    ax[1].set_xlabel("y coordinate in ckpc ")


    ax[2].set_xticks(np.arange(0,8e4,1.5e4))
    ax[2].hist(z, bins=20)
    ax[2].set_xlabel("z coordinate in ckpc ")


    plt.tight_layout()
    plt.savefig(name_string)

    x_p.append((x.min(), x.max()))
    y_p.append((y.min(), y.max()))
    z_p.append((z.min(), z.max()))
    d1.close()
```
